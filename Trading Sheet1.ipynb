```python
data = bt.get('aapl,goog', start='2010-01-01')
```

```python
bt_strategy = bt.Strategy('Trade_Weekly',
                         [bt.algos.RunWeekly(),
                          bt.algos.SelectAll(),
                          bt.algos.WeighEqually(),
                          bt.algos.Rebalance()])
```

```python
# Create a backtest
bt_test = bt.Backtest(bt_strategy, data)
```

```python
# Run the backtest
bt_res = bt.run(bt_test)
```

```python
bt_res.plot(title="Backtest result")
plt.show()
```
