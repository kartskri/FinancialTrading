```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bitcoin_data = pd.read_csv("Bitcoin.csv", index_col="Date", parse_dates=True)
bitcoin_data.head()
```

```python
plt.figure(figsize=(16,8))

# High Price Plot
plt.plot(bitcoin_data['High'], color="red", linestyle="dashed")

# Low Price Plot
plt.plot(bitcoin_data['Low'], color="green", linestyle="solid")



plt.show()
```
