```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
goog_data = pd.read_csv('GOOG.csv', index_col='Date', parse_dates=True)
goog_data
```

```python
goog_data['sma_50'] = goog_data['Close'].rolling(50).mean()
```

```python
plt.figure(figsize=(12,8))
plt.plot(goog_data['sma_50'], color='green', label='SMA_50')
plt.plot(goog_data['Close'], color='red', label='Close')
plt.title("Simple Moving Averages")
plt.show()
```
